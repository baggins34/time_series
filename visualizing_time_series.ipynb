```python
import pandas as pd
from matplotlib import dates
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Data/starbucks.csv", index_col="Date", parse_dates=True)
```

if the date column in strings and Pandas doesn't recognize the date;

```python
df.index = pd.to_datetime(df.index)
```

```python
df.plot(figsize=(16,6))
```

```python
ax = df['Close'].plot(figsize=(16,6), title='Closing')
ax.autoscale(axis='both', tight=True)
ax.set(xlabel="Time", ylabel='Closing')
```

```python
df['Close']['2017-01-01':'2017-12-31'].plot(figsize=(16,6))
```

```python
df['Close']['2017'].plot(figsize=(16,6))
```

```python
df['Close'].plot(figsize=(16,6), xlim=['2017-01-01','2017-12-31'])
```

```python
df['Close']['2017-01-01':'2017-12-31'].plot(figsize=(16,6), ls='--', c='green')
```

```python
ax = df['Close'].plot(xlim=['2017-01-01','2017-03-01'], ylim=[50,57], figsize=(16,6))
ax.set(xlabel='')

# Evenly and weekly spaced X axis
ax.xaxis.set_major_locator(dates.WeekdayLocator(byweekday=0)) # Here the 0 is monday
# Format of the X axis
ax.xaxis.set_major_formatter(dates.DateFormatter('%a-%B-%d-%Y'))

# Let's leave a makerer for each month
ax.xaxis.set_minor_locator(dates.MonthLocator())
ax.xaxis.set_minor_formatter(dates.DateFormatter('\n\n%b'))

ax.yaxis.grid(True)
ax.xaxis.grid(True)
```
